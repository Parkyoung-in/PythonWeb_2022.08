#HTML 파일 열어서 원하는 정보 가져오기

```python
from bs4 import BeautifulSoup
with open('example.html') as f:
  soup=BeautifulSoup(f, 'html.parser')
```

### 1. 태그 정보 : find(),find_all()

```python
# 첫번째 div 태그의 내용을 가져오기
first_div = soup.find('div')
first_div
```

```python
#해당되는 태그를 모두 가져오기
ps = first_div.find_all('p')
ps
```

```python
for ptag in ps:
  print(ptag)
```

```python
ps[1]
```

### 2. CSS Selector : select_one(), select()

```python
# class 명이 ex_class인 태그의 정보를 가져오기
second_div = soup.select_one('.ex_class')
second_div
```

```python
# id값이 ex_id인 태그의 정보를 가져오기
third_div = soup.select_one('#ex_id')
third_div
```

```python
# div 태그를 모두 가져오기 
all_divs=soup.select('div')
len(all_divs)
```

```python
# 모든 p 태그
all_ps = soup.find_all('p')
len(all_ps)
```

```python
# div의 자식인 p 태그
div_ps = soup.select('div p') #부자관계 앞(div)이 부 뒤(p)가 자식
len(div_ps)
```
