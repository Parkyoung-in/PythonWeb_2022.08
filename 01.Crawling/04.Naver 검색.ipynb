```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 화면없이 실행
options.add_argument('--no-sandbox')
options.add_argument('--single-process')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
# url에 한글을 입력하고 싶으면
from urllib.parse import quote
base_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&'
query = quote('스마트폰')
url = f"{base_url}query={quote('스마트폰')}"
driver.get(url)
time.sleep(1) #화면 바뀔때는 반드시 sleep을 걸어주어야 한다
```

```python
# 결과 화면에서 VIEW 클릭
driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/ul/li[6]/a').click()
```

```python
# Page Down 화살표를 사용하여 데이터를 충분히 가져오는 과정
body = driver.find_element(By.TAG_NAME, 'body')
for i in range(30):
  body.send_keys(Keys.PAGE_DOWN)
  time.sleep(2)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.close()
```
